```python
def mean_of_poverty (state_name, column_name):
    state_df= dem_data.loc[dem_data["State"]==state_name, [column_name]]
    return state_df[column_name].mean()
```

```python
mean_of_poverty('TX', 'Poverty')
```

```python
mean_of_poverty('AL','Poverty')
```

```python
mean_of_poverty('OK','Suicide')
```

```python
#creating a bad chart of mean of poverty of all states
bar_chart= pygal.Bar()
bar_chart.title= "Mean of Poverty for all States in USA"
for state in list (dem_data["State"].unique()):
    bar_chart.add(state, mean_of_poverty(state,'Poverty'))
SVG(bar_chart.render())
```

```python
#creating a bad chart of mean of poverty of all states - horizontal bars
bar_chart= pygal.HorizontalBar()
bar_chart.title= "Mean of Poverty for all States in USA"
for state in list (dem_data["State"].unique()):
    bar_chart.add(state, mean_of_poverty(state,'Poverty'))
SVG(bar_chart.render())
```

```python
#creating a bad chart of mean of poverty of all states - horizontal bars- adjusting legends
bar_chart= pygal.HorizontalBar(legend_at_bottom= True)
bar_chart.title= "Mean of Poverty for all States in USA"
for state in list (dem_data["State"].unique()):
    bar_chart.add(state, mean_of_poverty(state,'Poverty'))
SVG(bar_chart.render())
```
